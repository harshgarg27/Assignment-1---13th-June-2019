## Quesstion-2

### Plot 'Argentina-CABA' and 'Argentina-Catamarca' for number of cases values for entire duration to identify the difference

```python
df2=df.query('unit=="cases" & (location=="Argentina-Catamarca" | location=="Argentina-CABA")')
```

```python
## convertung value column to numeric type
df2['value'] = pd.to_numeric(df2.value)
df2.dtypes
```

```python
## Grouping by location first, and then grouping by report date 
## After this, we will get sum of values in values column for each date
new_df = df2.groupby('location')
df3=new_df.get_group("Argentina-Catamarca").groupby('report_date').sum()
df4=new_df.get_group("Argentina-CABA").groupby('report_date').sum()
```

```python
df4.head(3)
```

```python
df3.head()
```

```python
new_df=df3[df3.columns[-1]] ##catamarca
new_df2=df4[df4.columns[-1]]  ##Caba
```

```python
report_date=new_df.index.tolist()
```

```python
barwidth = 0.25
r1 = np.arange(len(new_df))
r2 = [x + barwidth for x in r1]

plt.style.use('seaborn-whitegrid')
##Bar
plt.bar(r1,new_df2,width=barwidth,label='Argentina-Catamarca',color='darkred')
plt.bar(r2,new_df,width=barwidth,label='Argentina-CABA',color='darkgreen')

##Line
plt.plot(report_date,new_df,"g-.",label="Argentina-CABA",color="magenta")
plt.plot(report_date,new_df2,"g--",label="Argentina-Catamarca",color="darkblue")

plt.xlabel('Dates', fontsize = 15, weight = 'bold')
plt.ylabel('sum of value', fontsize = 15, weight = 'bold')
plt.rcParams['font.family'] = "sans-serif"

plt.xticks([r + barwidth for r in range(len(new_df))])
plt.xticks(rotation=90)
plt.legend(loc=2)
plt.show()
```

## 3. Plot how data field values are distributed across the locations 'Argentina-Buenos_Aires', 'Argentina-CABA', 'Argentina-Catamarca'

Solution:

```python
## First filtering out only  'Argentina-Buenos_Aires', 'Argentina-CABA', 'Argentina-Catamarca' from dataframe

loc_df = df[(df.location=='Argentina-Catamarca') | (df.location=='Argentina-Buenos_Aires') | (df.location=='Argentina-CABA')]
```

```python
loc_df.head()
```

```python
loc_df['value'] = pd.to_numeric(loc_df.value)
```

```python
data_frame=loc_df.groupby('location')
```

```python
catamarca = data_frame.get_group('Argentina-Catamarca').groupby('data_field')['value'].sum()
```

```python
catamarca
```

```python
caba = data_frame.get_group('Argentina-CABA').groupby('data_field')['value'].sum()
```

```python
caba
```

```python
beunos = data_frame.get_group('Argentina-Buenos_Aires').groupby('data_field')['value'].sum()
```

```python
beunos
```

```python
plt.subplots(figsize=(60,20))
colors = ['#ff9999','#66b3ff','#99ff99','#BC8F8F','#ffcc99']
plt.subplot(3,1,1)
plt.title('Argentina-Buenos_Aires')
labels = beunos.index.tolist()[:5]
sizes = beunos.tolist()[:5]
pie_properties = plt.pie(sizes,shadow=True,autopct='%1.1f%%',startangle=90,colors=colors,labeldistance=0.7,rotatelabels = 90,pctdistance=0.5)
plt.legend(labels, loc="best")
fraction_text_list = pie_properties[2]
for text in fraction_text_list:
    text.set_rotation(90)


plt.subplot(3,1,2)
plt.title('Argentina-Catamarca')
labels = catamarca.index.tolist()[:5]
sizes = catamarca.tolist()[:5]
pie_properties=plt.pie(sizes,shadow=True,autopct='%1.1f%%',startangle=90,colors=colors,labeldistance=0.7,pctdistance=0.5,rotatelabels = 90)
plt.legend(labels, loc="best")
fraction_text_list = pie_properties[2]
for text in fraction_text_list:
    text.set_rotation(90)

plt.subplot(3,1,3)
plt.title('Argentina-CABA')
labels = caba.index.tolist()[:5]
sizes = caba.tolist()[:5]
pie_properties=plt.pie(sizes,shadow=True,autopct='%1.1f%%',startangle=90,colors=colors,labeldistance=0.7,pctdistance=0.5,rotatelabels = 90)
plt.legend(labels, loc="best")
fraction_text_list = pie_properties[2]
for text in fraction_text_list:
    text.set_rotation(90)
plt.show()
```
